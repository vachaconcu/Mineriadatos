# Red Neuronal Artificial 

Una Red Neuronal Artificial o ANN (Artificial Neuronal Network) por sus siglas en inglés, es un modelo computacional inspirado en una reurona biológica. Para ilustar, considérese la matriz

\begin{align*}
D = \begin{pmatrix} x_1 & y_1 \\ x_2 & y_2 \\ \vdots & \vdots \\ x_n & y_n\end{pmatrix}
\end{align*}

donde cada $x_i \in \mathbb{R}^d$ y cada $y_i \in \mathbb{R}$.  La predicción o $\hat{y_i}$ puede ser obtenido mediante la función

\begin{align*}
g_p= \left\{ \begin{array}{lcc} 1 &  \text{si}  & w^tx_i > 0 \\ 0 &  \text{en otro caso} \end{array}\right.
\end{align*}

la representación de la Fig 3.2 también es conocido como $\textit{perceptrón}$.

Ahora, no consideraremos el caso dicotómico, sino cuando se tiene más de dos clases, supongamos $\textbf{c}$. Consideremos también la matriz $\textbf{W}$

\begin{align*}
\textbf{W} = \begin{pmatrix} w_0^{(1)} & w_0^{(2)} & \cdots & w_0^{(c)} \\ w_1^{(1)} & w_1^{(2)} & \cdots & w_1^{(c)} \\ \vdots & \vdots \\ w_d^{(1)} & w_d^{(2)} & \cdots & w_d^{(c)} \end{pmatrix}
\end{align*}

\begin{align*}
\textbf{l}_1 = [l_1^{(1)}, l_1^{(2)}, ... , l_1^{(c)}], \hspace{0.5cm} \text{donde}
\end{align*}

\begin{align*}
\textbf{l}_1 = W^Tx_i
\end{align*}

el subíndice 1 de $\textbf{l}_1$ hace referencia a la salida de la primera capa. Ahora, se define la función de activación $\textbf{softmax-activated}$ como:

\begin{align*}
P(y_i|x_i) = \sigma(l_1^{(j)}) = \displaystyle\frac{e^{l_1^{(j)}}}{\displaystyle\sum_{k=1}^{c}e^{l_1^{(j)}}}, \hspace{0.5cm} j=1,...,c
\end{align*}

Finalmente la clase predicha es la clase que maximiza la anterior probabilidad, es decir

\begin{align*}
\hat{y_i} = argmax_j P(y_i=j|x_i)
\end{align*}

# Convoluciones

Las convoluciones son un procedimiento por el cual una imagen es reducida a un espacio compacto o de menor dimensión en el momento en que:

* Existe correlación en los datos. Por ejemplo en imágenes, la correlación puede existir entre pixeles y para audios, la correlación puede estar presente en la variable tiempo.
* Mejorar el rendimiento del modelo. En la práctica no se recomienda que una neurona muy estructurada o pesada se mueva alrededor de la señal, en cambio, genera más rendimiento el recorrido de varias neuronas más ligeras que comparten estructuras de pesos.
* Cuando no existen pesos para una componente de un vector de entrada, se toman pesos adyacentes en lugar de añadir un nuevo parámetro.

Una convolución puede verse como un proceso de reducción de dimensionalidad representado en el uso de una cantidad de pesos menor a la dimensión de los vectores de entrada, es decir:

Sea **x** = [x^(1), x^(2), ... , x^(d)] un vector de entrada. Teniendo un conjunto de pesos de orden fijo

\begin{align*}
w = [w_1, w_2, ...,w_k]
\end{align*}

se define una **Convulución** como la operación

\begin{align*}
z^{(j)} = \displaystyle\sum_{i=1}^k x^{(j+i-1)}w_k, \hspace{0.5cm} j=1,2, ... ,d-k+1
\end{align*}

Por ejemplo, las componentes del nuevo vector **z** serán:

\begin{align*}
z^{(1)} &= x^{(1)}w_1 + x^{(2)}w_2 + ... + x^{(k)}w_k\\
z^{(2)} &= x^{(2)}w_1 + x^{(3)}w_2 + ... + x^{(k+1)}w_k\\
z^{(3)} &= x^{(3)}w_1 + x^{(4)}w_2 + ... + x^{(k+2)}w_k\\
\vdots \\
z^{(d-k+1)} &= x^{(d-k+1)}w_1 + x^{(d-k+2)}w_2 + ... + x^{(d)}w_k\\
\end{align*}

obteniendo finalmente un vector de dimensión *d-k+1* con k-pesos, donde $k<d$.

# Redes Discriminatorias

Las redes discriminatorias son aquellas en las cuales se intenta minimizar la pérdida de predicción de una distribución sobre las etiquetas de categorización de interés. Estas redes muestran datos de alta dimensión como imágenes y los asignan a un espacio con representaciones características.

## Redes Generativas

Son una red discriminatoria particular que toman una muestra de los datos asociados a una distribución en un espacio generalmente más pequeño que el espacio de los datos y generan muestras que imitan las muestras extraídas de las distribuciones de mayor dimensión.

## Autoencoder

Los autoencoder son redes que aprenden a regenerar la imagen de manera propia. Este estudio consta de dos partes: $\textit{Enconder}$ y $\textit{Decoder}$, primero se crea una representación resumida y compacta de la imagen, y luego se construye la imagen a partir de estas imágenes compactas. En la literatura, a estas redes regenerativas reciben el nombre de redes $\textit{simétricas}$ ya que generalmente el decoder usa los mismos pesos que el encoder solo que traspuestos, por ejemplo, al usar convuluciones los filtros del decoder son los traspuestos del encoder. Considerando un vector $\textbf{x}$, el encoder $E(\textbf{x})=\textbf{c}$ toma dicha entrada y produce una representación $\textbf{c}$ la cual es usada por el decoder $D(\textbf{c})=\hat{x}$. Ya los pesos del encoder y el decoder están ligados, basta con encontrar algo de los dos; el método que por lo general se usa es el RMSE (Root Mean Squeare Error). Adicionalmente se debe mencionar que es muy común usar un autoencoder como inicialización de redes discriminatorias.
